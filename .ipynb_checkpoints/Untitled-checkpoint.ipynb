```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Ahmedabad_weather.csv')
```

```python
data = data.drop(['preciptype','severerisk'], axis=1)
```

```python
data = data.drop(['name','sunrise','sunset','conditions','description','icon','stations'],axis=1)
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
```

```python
data['tempdiff'] = data['tempmax']-data['tempmin']
```

```python
data.groupby('month')['temp'].mean().plot(kind='line')
plt.title('Avg Temprature by Month')
plt.show()
```

```python
data.groupby('month')['windspeed'].mean().plot(kind='line')
plt.title('Average Windspeed by Month')
plt.show()
```

```python
train_size = int(len(data)*0.8)
train,test = data[:train_size], data[train_size:]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train[['tempmin','windspeed','humidity']],train['tempmax'])
predictions = model.predict(test[['tempmin','windspeed','humidity']])
```
