```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('WeatherData.csv')
```

```python
data = data.drop(['preciptype','severerisk'], axis=1)
```

```python
data = data.drop(['name','sunrise','sunset','conditions','description','icon','stations'],axis=1)
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
```

```python
data['tempdiff'] = data['tempmax']-data['tempmin']
```

```python
data['prevdaytemp'] = data['temp'].shift(1)
data['prevdaytempmax'] = data['tempmax'].shift(1)
data['prevdaytemp'] = data['prevdaytemp'].fillna(data['temp'])
data['prevdaytempmax'] = data['prevdaytempmax'].fillna(data['tempmax'])
```

```python
train_size = int(len(data)*0.5)
train,test = data[:train_size], data[train_size:]
```

```python
model = LinearRegression()
columns = ['prevdaytemp','prevdaytempmax']
model.fit(train[columns],train['tempmax'])
realday = pd.DataFrame([[29.0, 31.49]], columns=columns)
realday_predict = model.predict(realday)
formatted_prediction = f"{realday_predict[0]:.2f}째C"
print(f"Today's Maximum Temprature Prediction based on Yeterdays's Temprature and  Yeterdays's Maximum Temprature:{formatted_prediction}")


columns = ['prevdaytemp','windspeed']
model.fit(train[columns],train['temp'])
realday = pd.DataFrame([[29.92, 11.0]], columns=columns)
realday_predict = model.predict(realday)
formatted_prediction = f"{realday_predict[0]:.2f}째C"
print(f"Today's Temprature Prediction based on Yeterdays's Temprature and Today's Windspeed:{formatted_prediction}")
```

```python
predictions = model.predict(test[columns])
plt.scatter(test['temp'], predictions)
plt.plot([min(test['temp']), max(test['temp'])], [min(test['temp']), max(test['temp'])], color='red')  # 45-degree line
plt.title("Predicted vs. Actual Temperature")
plt.xlabel("Actual Temperature")
plt.ylabel("Predicted Temperature")
plt.show()
```

```python

mape = np.mean(np.abs((test['temp'] - predictions) / test['temp'])) * 100
print(f"MAPE: {mape}%")
```
